```python
import numpy as np
import pandas as pd
import os
import re
```

```python
data1 = pd.read_excel('相似度.xlsx',encoding='gbk')
data2 = pd.read_csv('增加可解释性词.csv',  encoding = 'GB18030')
data_temp = pd.merge(data1, data2, on='留言编号')                          #横向拼接

data3 = pd.read_excel('回复时间.xlsx',encoding='gbk')
data_new = pd.merge(data_temp, data3, on='留言编号')
data_new.to_excel('准备计分.xlsx',encoding = 'GB18030')
```

```python
score = pd.read_excel('准备计分.xlsx',encoding='gbk',index=0)
score
```

```python
def first_end(x):
    if x == '完整':
        a = 0.3
    else:
        a = 0
    return a
```

```python
def complete(x):
    if x == '已解决':
        a = 0.4
    else:
        a = 0
    return a
```

```python
def conj(x):
    if len(x)> 2:
        a = 1
    else:
        a = 0
    return a
```

```python
score['f_score']=score['f_first'].apply(lambda x: first_end(x) )
score['e_score']=score['f_end'].apply(lambda x: first_end(x) )
score['com_score']=score['f_complete'].apply(lambda x: complete(x) )
score['conj_score']=score['可解释性词'].apply(lambda x: conj(x) )
```
