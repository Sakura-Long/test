```python
answer=pd.DataFrame(df.loc[:,['留言编号','答复意见']])
answer
```

```python
answer['答复意见'] = answer['答复意见'].apply(lambda x : re.sub('\n','',str(x)))
answer['答复意见'] = answer['答复意见'].apply(lambda x : re.sub('\t','',str(x)))
answer['答复意见'] = answer['答复意见'].apply(lambda x : re.sub('\r','',str(x)))
answer['答复意见'] = answer['答复意见'].apply(lambda x : re.sub('\s','',str(x)))
answer['答复意见'] = answer['答复意见'].apply(lambda x : re.sub('http[A-Za-z:/.]+','',str(x)))
answer
```
